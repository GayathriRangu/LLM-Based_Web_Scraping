```python
from bs4 import BeautifulSoup
import requests
page=requests.get("https://en.wikipedia.org/wiki/Deep_learning") #get the page
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify()) #prettify the content of the page
```

```python
list(soup.children) #list of all the children of the page
```

```python
print(soup.find_all('p')) #find all the paragraph tags in the page
print('\n\n') #print them line by line
```

```python
print(soup.find_all('p')[0].get_text()) #get the text of the first paragraph tag
```

2. SCRAPING THE WIKI TABLE WITH PANDAS USING read_html()

read_html() is used to read the HTML tables directly froma  webpage into a pandas data frame without needing to scrape the website's html explicitly. 

```python
import pandas as pd
html_string ='''
<table>
<tr>
<th>Header 1</th>
<th>Header 2</th>
</tr>
<tr>
<td>row 1, cell 1</td>
<td>row 1, cell 2</td>
</tr>
<tr>
<td>row 2, cell 1</td>
<td>row 2, cell 2</td>
</tr>
</table>
'''
df_1=pd.read_html(html_string)
df_1
```
